### Задание 1

В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```
### YOUR CODE/SOLUTION HERE
goods1 = 0.2
goods2 = 0.3
goods3 = 0.5

prem_goods1 = 0.1
prem_goods2 = 0.05
prem_goods3 = 0.2

#Событие Аi - вероятность купить любую продукцию конкретного предприятия (см.goods1,2,3)
#Событие Bi - вероятность купить премиум продукцию конкретного предприятия (напр.: goods1 * prem_goods1)
#События Bi-ые независимы между собой, поэтому вероятности складываем

p_prem = (goods1*prem_goods1 + goods2*prem_goods2 + goods3*prem_goods3)*100
print(f'Вероятность, что купленная новая продукция окажется высшего сорта {p_prem}%')
```

### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```
### YOUR SOLUTION HERE

# Доказательство попарной независимости

'''
Два события считаются независимыми, если вероятность их совместного наступления
равна произведению их индивидуальных вероятностей
'''

'''
Если сторона монеты не важна, то для двух монет возможны 2 исхода:
    1) Выпали Орел-Орел / Решка-Решка
    2) Выпали Орел-Решка / Решка-Орел
Вероятность исхода 1) равна 1/2, вероятность исхода 2) равна 1/2.
'''

Pi_1 = 1/2 # вероятность, что выпадет конкретная сторона (напр. Решка) монеты 1
Pi_2 = 1/2 # вероятность, что выпадет конкретная сторона (напр. Решка) монеты 2
Pi_3 = 1/2 # вероятность, что выпадет конкретная сторона (напр. Решка) монеты 3

P_A1i = Pi_1 * Pi_2 #1/4 - вероятность, что и 1, и 2 монета упали одной и той же конкретной стороной
P_A1 = P_A1i + P_A1i # 1/2 вероятность, что после подброса у монет 1 и 2 одна сторона (неважно какая)

P_A2i = Pi_2 * Pi_3 #1/4 - вероятность, что и 2, и 3 монета упали одной и той же конкретной стороной
P_A2 = P_A2i + P_A2i # 1/2 вероятность, что после подброса у монет 2 и 3 одна сторона (неважно какая)

P_A3i = Pi_1 * Pi_3 #1/4 - вероятность, что и 1, и 2 монета упали одной и той же конкретной стороной
P_A3 = P_A3i + P_A3i # 1/2 вероятность, что после подброса у монет 1 и 2 одна сторона (неважно какая)


#Независимость А1 и А2
if P_A1*P_A2 == P_A1*P_A2:
  print('События A1 и A2 независимые')
else:
  print('События A1 и A2 зависимые')

#Независимость А2 и А3
if P_A2*p_A3 == P_A2*p_A3:
  print('События A2 и A3 независимые')
else:
  print('События A2 и A3 зависимые')

#Независимость А1 и А3
if P_A1*p_A3 == P_A1*p_A3:
  print('События A1 и A3 независимые')
else:
  print('События A1 и A3 зависимые')

```

```
# Доказательство отсуствия независимости событий в совокупности
'''
Если сторона монеты важна, то для двух монет возможны 4 исхода:
    1) Выпали Орел-Орел
    2) Выпали Орел-Решка
    3) Выпали Решка-Решка
    4) Выпали Решка-Орел
Вероятность каждого из исходов равна 1/4 (т.к. 1/2 * 1/2)
'''


Pi_1 = 1/2 # вероятность, что выпадет конкретная сторона (напр. Решка) монеты 1
Pi_2 = 1/2 # вероятность, что выпадет конкретная сторона (напр. Решка) монеты 2
Pi_3 = 1/2 # вероятность, что выпадет конкретная сторона (напр. Решка) монеты 3

#Шаг 1
P_A1i = Pi_1 * Pi_2 #1/4 - вероятность, что и 1, и 2 монета упали одной и той же конкретной стороной
P_A1 = P_A1i + P_A1i # 1/2 вероятность, что после подброса у монет 1 и 2 одна сторона (неважно какая)

#Шаг 2
#для события А2 уже важно, какой стороной выпала монета 2 в событии А1
P_A2 = (Pi_2*Pi_2) * Pi_3 #1/8 - вероятность, что и 2, и 3 монета упали одной и той же конкретной стороной (при условии А1)

#Шаг 3
#для события А3 уже важно, какой стороной выпала монета 1 в событии А1 и монета 3 в событии А2
P_A3 = (Pi_1*Pi_1) * (Pi_3 * Pi_3)
#1/16 - вероятность, что и 1, и 3 монета упали одной и той же конкретной стороной,
# или вероятность наступления события А3, при условии, что события А1 и А2 наступили


#Независимость А1,А2,А3 в совокупности (P(A1A2A3)==P(A1)*P(A2)*P(A3))
if P_A3 == P_A1*P_A2*P_A3:
  print('События А1,А2,А3 независимые в совокупности')
else:
  print('События А1,А2,А3 зависимые в совокупности')
```

**Нормальное распределение**

```
norm_rv = stats.norm(loc=5, scale=2)
x=np.linspace(0,10,100) #генерируем набор точек по x и значения функций
norm_rv_pdf = norm_rv.pdf(x)
plt.plot(x, norm_rv_pdf)
plt.show()
```

У нормального распределения:

1) Значение loc (среднее значение) влияет на то, в какую сторону будет смещена функция. В нашем случае, если ближе к 0, то влево, если ближе к 10 - вправо.

2) Значение scale (разброс, дисперсия) влияет на ширину функции и ее "хвосты". Чем больше значение scale, тем функция шире, а её "хвосты" короче.
