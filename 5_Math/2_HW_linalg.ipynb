### Задание 1

Дан объект в $2D$ пространстве

````
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````
#Создаем линейный оператор, который уменьшит исходную матрицу на 2
B = np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 0.5]])
B
````

````
#С помощью цикла перебираем координаты точек А и задаем им новые значения
tt=[]
for kk in A:
    tt.append(B@kk)
````

````
tt
````

````
A=np.array(tt)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
#Повернем объект на 130 градусов (по часовой стрелке)
import math
B2=[[math.cos(130*math.pi/180), math.sin(130*math.pi/180), 0],
 [-(math.sin(130*math.pi/180)), math.cos(130*math.pi/180),0],
  [0, 0, 1]]
````

````
B2=np.array(B2)
res1=B2@A.transpose()
res2=res1.transpose()
````

````
A=np.array(res2)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
#Отобразим объект относительно прямой y=x согласно условию
A
````

````
K=np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
````

````
res_3=K@A.transpose()
res_4=res_3.transpose()
````

````
res_4
````

````
A = np.array(res_4)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````
