```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
'''
Вопрос 1:
Какой контент (по категориям) наиболее популярный?
'''

sns.countplot(data=df_youtube, y="category", hue='category',  legend=False)
plt.title("Количество каналов в разрезе категорий")
plt.ylabel("Категория канала")
plt.xlabel("Количество категорий")
plt.show()

#Больше всего каналов среди развлекательного контента: развлечения, музыка
```

```python
'''
Вопрос 2:
Каналы из каких стран встречаются чаще всего (топ 5)?
'''

sns.barplot(data=df_youtube['Country'].value_counts().to_frame().reset_index().loc[:4], x='Country', y="count", hue='Country',  legend=False)
plt.title("Количество каналов в разрезе стран (топ 5)")
plt.ylabel("Количество каналов")
plt.xlabel("Страна")
plt.show()

#Больше всего каналов принадлежат пользователям из США и Индии
```

```python
'''
Вопрос 4:
Какой контент (по просмотрам) был наиболее популярным за поледний месяц?
'''

df_youtube.groupby('category')['video_views_for_the_last_30_days'].sum().astype(int).sort_values().plot(kind="bar", )
plt.title("Количество просмотров контента за последний месяц")
plt.ylabel("Количество просмотров")
plt.xlabel("Категория")
plt.show()

#Топ 3 категорий по просмотрам за последний месяц примерно такой же, что и топ 3 категорий по просмотрам в целом.
#За последний месяц бОльший интерес наблюдается к категории "Entertainment"
#Категория "Education" не вошла в топ 5 (в отличие от данных графика к вопросу 3)
```

```python
'''
Вопрос 5:
Каналы каких категорий содержат больше всего видео/выпусков?
'''

df_youtube.groupby('category')['uploads'].sum().astype(int).sort_values().plot(kind="bar", )
plt.title("Количество видео на каналах")
plt.ylabel("Количество видео")
plt.xlabel("Категория")
plt.show()

#Авторы новостных и развлекательных каналов больше остальных пополняют контент (загружают видео/выпуски).
```

```python
'''
Вопрос 6:
Есть ли связь между уровнем безработицы и количеством просмотров (за последний месяц)?
'''

df_youtube.plot.scatter(x='Unemployment rate',
                        y='video_views_for_the_last_30_days',
                        title='Связь между уровнем безработицы и количеством просмотров (за последний месяц)',
                        xlabel='Уровень безработицы',
                        ylabel='Количество просмотров за последний месяц')

#Как видно из графика, более широкий размах по просмотрам наблюдается при безработице на уровне 4-6% и >14%
#Если положительная корреляционная связь и есть, то, возможно, она незначительная.
```

Как изменится точность, если опустить пункт 2 и не убирать из данных нулевые просмотры?




```python
#Оставляем два коррелирующих столбца: "subscribers" и"video views".
df_youtube_2 = df_youtube[['subscribers', 'video views']]
```

```python
df_youtube_2.info()
```

```python
#Убираем те строки, где кол-во просмотров равно нулю.

df_youtube_2 = df_youtube_2[(df_youtube_2['subscribers'] != 0) & (df_youtube_2['video views'] !=0)]
```

```python
df_youtube_2.info()
```

```python
#Рассчитываем коэффициент корреляции Пирсона.
from scipy.stats import pearsonr
```

```python
corr_pearson = pearsonr(df_youtube_2['subscribers'], df_youtube_2['video views'])

print(f'Коэффициент корреляции Пирсона: {corr_pearson[0]:.4}')
print(f'p-value: {corr_pearson[1]:.4f}')
```

Коэффициент корреляции 0.8243 свидетельствует о высокой положительной связи между двумя переменными.

Значение p-value меньше 0,05 говорит о том, что связь между переменными статистически значимая.

```python
#Определяем "video views" как признак Х, а "subscribers" - как целевую переменную.
X = df_youtube_2[['video views']]
y = df_youtube_2['subscribers']
```

```python
#Разделяем выборку на тестовую и тренировочную.
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
#Строим модель линейной регрессии.
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
res = model.fit(X_train, y_train)
res
```

```python
y_pred = model.predict(X_test)
y_pred #Предсказанное число подписчиков на основании модели
```

```python
model.score(X_test, y_test) # метод возвращает значение коэффициента детерминации

#Точность модели слабая
```

```python
'''
Посмотрим, как изменится точность модели, если использовать все строки
'''
df_youtube_full = df_youtube[['subscribers', 'video views']]
```

Изменится ли точность модели, если в качестве признаков выбрать другие столбцы?

```python
#Определим две самые популярные категории канала и отфильтруем по ним набор данных.
df_youtube['category'].value_counts().sort_values(ascending=False).iloc[:2]
```

```python
category_top2 = df_youtube.loc[df_youtube['category'].isin(['Entertainment', 'Music'])]
category_top2.info()
```

```python
#В качестве признаков используем следующие столбцы: [‘video views’, ‘subscribers’, ‘uploads’]
df_category_top2 = category_top2[['category', 'video views', 'subscribers', 'uploads']].reset_index(drop=True)
```

```python
df_category_top2.head()
```

```python
#Приведем значения столбца 'video views' к типу данных int
df_category_top2['video views'] = df_category_top2['video views'].astype(int)
df_category_top2.head()
```

```python
#Уберем те строки, где кол-во просмотров, загрузок равно нулю.

df_category_top2 = df_category_top2[(df_category_top2['video views'] != 0) & (df_category_top2['uploads'] != 0)]
df_category_top2.info()
```

```python
#Для целевой переменной используем столбец 'category'.
#Значения по нему закодируем с помощью LabelEncoder.
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

df_category_top2.loc[:, 'le_category'] = le.fit_transform(df_category_top2.loc[:, 'category'])

'''
После кодировки (cм.le_category):
  Entertainment - 0
  Music - 1
'''
```

```python
df_category_top2.drop(('category'), axis=1, inplace=True)
df_category_top2.info()
```

```python
df_category_top2.head()
```

```python
X = df_category_top2.iloc[:, 0:3]
```

```python
y = df_category_top2['le_category']
```

```python
#Разделим выборку на тестовую и тренировочную, укажем random_state=13.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=13)
```

```python
#построим модель логистической регрессии.
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
predictions[:5] # посмотрим как выглядят наши предсказания
```

```python
#Рассчитаем точность модели на тестовой выборке.
model.score(X_test,y_test)
```
