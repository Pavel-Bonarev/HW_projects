**Задание 1. Загрузка данных**

Изучить представленный набор данных на основе описания его столбцов в файле “horse_data.names” , загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/horse_data.csv')
df.head()
# Проблема: в заголовки столбцов попали значения
```

**Задание 3. Работа с пропусками**

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
#Находим количество пропусков в каждом столбце
df_for_work.isnull().sum()
```

```python
df_for_work.isna().mean() * 100 #Находим долю пропусков в столбцах
```

Значение в temperature of extremities коррелируют со значениями в rectal temperature, в связи с чем пропуски в rectal temperature следует заполнить средним значением температуры в разрезе каждой группы temperature of extremities

```python
print(df_for_work.groupby('temperature of extremities')['rectal temperature'].mean())
df_for_work['rectal temperature'] = df_for_work['rectal temperature'].fillna(df_for_work.groupby('temperature of extremities')['rectal temperature'].transform('mean'))
```

```python
#Столбцы surgery и outcome почти не содержат пропусков, пропуски можем заменить на значение, которое не входит в категорию
df_for_work['surgery'] = df_for_work['surgery'].fillna('0')
df_for_work['outcome'] = df_for_work['outcome'].fillna('0')
```

```python
#Столбец Age содержит лишь категрию 1 и пропуски, категории 2 (молодые лошади) в значениях нигде нет, в связи с чем
#можем сделать предположение, что при заполнении данных на молодых лошадей пропуск в графе Age сделан намеренно
#пропуски следует заполнить категорией 2
df_for_work['Age'] = df_for_work['Age'].fillna('2')
```

```python
#Значения в столбце respiratory rate малополезны, поскольку подвержены большим колебаниям
#Можем заменить на среднее
df_for_work['respiratory rate'] = df_for_work['respiratory rate'].fillna(df_for_work['respiratory rate'].mean())
```

```python
#Поскольку чем выше уровень боли, тем выше вероятность операции, то можем заполнить столбец pain наиболее встречающимся значением в разрезе категорий столбца surgery
print(df_for_work.groupby('surgery')['pain'].agg(pd.Series.mode))
```

```python
def change_pain(row):
    if row['surgery'] == '0':
        return '4'
    elif row['surgery'] == '1':
        return '3'
    else:
        return '1'
```

```python
df_for_work['pain']=df_for_work.apply(change_pain, axis=1)
```

```python
#Учащенный пульс может быть связан с уровнем боли
print(df_for_work.groupby('pain')['pulse'].mean())
df_for_work['pulse'] = df_for_work['pulse'].fillna(df_for_work.groupby('pain')['pulse'].transform('mean'))
```

```python
print(df_for_work.groupby('surgery')['pain'].agg(pd.Series.mode))
```

```python
#столбец rectal temperature имеет небольшое количество пропусков, в связи с чем пропуски можем заменить на среднее значение
df_for_work['rectal temperature'] = df_for_work['rectal temperature'].fillna(df_for_work['rectal temperature'].mean())
```

```python
#Часть пропусков в temperature of extremities можно заполнить категорией 2, если температура лошади повышенная
```

```python
def temp(row):
    if row['rectal temperature'] >= 37.8:
        return '2'
    else:
        return row['rectal temperature']
```

```python
df_for_work['temperature of extremities']=df_for_work.apply(temp, axis=1)
```
